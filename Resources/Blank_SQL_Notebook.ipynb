```python
%%sql

SELECT orderdate,
  quantity * unitprice AS total_price,
  c.customerkey,
  c.gender,
  c.latitude,
  c.longitude
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
WHERE orderdate::date BETWEEN '2024-01-01' AND '2024-04-30'
```

```python
%%sql

SELECT
  SUM(quantity * unitprice)AS total_price,
  c.gender
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
WHERE orderdate::date BETWEEN '2024-01-01' AND '2024-04-30'
GROUP BY gender
```

```python
%%sql

SELECT gender,
  COUNT(gender) AS gender_count
FROM customer
GROUP BY gender
```
